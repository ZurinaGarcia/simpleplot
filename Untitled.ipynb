```python
# primero cargamos toda la info

import os, geopandas as gpd

countries=gpd.read_file(os.path.join("maps","World_Countries","World_Countries.shp"))
rivers=gpd.read_file(os.path.join("maps","World_Hydrography","World_Hydrography.shp"))
cities=gpd.read_file(os.path.join("maps","World_Cities","World_Cities.shp"))
```

```python
# Vista del mapa
countries.plot(facecolor="lavenderblush",
               edgecolor='black',
               linewidth=0.1)
```

```python
# Vista de los ríos
rivers.plot(edgecolor='steelblue', 
            linewidth=1,
            linestyle='dotted')
```

```python
# Vista de las ciudades

cities.plot(marker='.', # marker type
            color='crimson', 
            markersize=1,
            alpha=0.3) 
```

```python
# como todos presentan el mismo CRS, continuamos

base = countries.plot(facecolor="lavenderblush", edgecolor='black', linewidth=0.1,figsize=(12,12))
cities.plot(marker='.', color='crimson', markersize=1,alpha=0.7,
            ax=base)
rivers.plot(edgecolor='steelblue', linewidth=0.4,
            ax=base)
```

```python
import folium


m = cities.explore(color="red", 
                   name="cities")

m = rivers.explore(m=m, color="blue",
                   name="rivers")
m
```

```python
# Elegimos un país con el que trabajaremos

ukraine=countries[countries.COUNTRY=='Ukraine']
```

```python
ukraine.plot()
```

```python
cities_ukraine=cities[cities.COUNTRY=='Ukraine']
cities_ukraine.plot()
```

```python
rivers_ukraine = gpd.clip(gdf=rivers,
                               mask=ukraine)

rivers_ukraine.plot()
```

```python
base = ukraine.plot(facecolor="tan", edgecolor='black', linewidth=0.4,figsize=(5,5))
cities_ukraine.plot(marker='+', color='brown', markersize=15,ax=base)
rivers_ukraine.plot(edgecolor='teal', linewidth=0.5,ax=base)
```

```python
# Mapa interactivo

ukraineCoords=[48.379433, 31.16558]
```

```python
m = cities_ukraine.explore(location=ukraineCoords,
                   zoom_start=4,
                   tiles='CartoDB positron',
                   color='red',
                   name="cities") 
m = rivers_ukraine.explore(m=m, color="blue",
                   name="rivers")

m
```

```python
ukraine_5641=ukraine.to_crs(5641)
cities_ukraine=cities[cities.COUNTRY=='Ukraine']
cities_5641=cities_ukraine.to_crs(ukraine_5641.crs)

rivers_5641=rivers.to_crs(ukraine_5641.crs)
rivers_5641=gpd.clip(gdf=rivers_5641,mask=ukraine_5641)
```

```python
base = ukraine_5641.plot(facecolor="t", edgecolor='black', linewidth=0.4,figsize=(5,5))
cities_5641.plot(marker='+', color='red', markersize=15,ax=base)
rivers_5641.plot(edgecolor='blue', linewidth=0.5,ax=base)
```
